```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#  read json file into a pandas dataframe
pareto_random_op_10_df = pd.read_json('trace_files/op-10/pareto_trace.log_waf.json')
pareto_random_op_15_df = pd.read_json('trace_files/op-15/pareto_trace.log_waf.json')
sequential_op_10_df = pd.read_json('trace_files/op-10/seq_trace.log_waf.json')
sequential_op_15_df = pd.read_json('trace_files/op-15/seq_trace.log_waf.json')
sequential_op_15_df
```

```python
# plot the data and show the max value of waf for each trace

fig, ax = plt.subplots()
ax.set_title('WAF vs Time')
ax.set_xlabel('Sequence Number')
ax.set_ylabel('WAF')

# ax.plot(sequential_op_15_df['waf'], label='sequential_op_15')
# ax.plot(sequential_op_10_df['waf'], label='sequential_op_10')
# ax.plot(pareto_random_op_15_df['waf'], label='pareto_random_op_15')
# ax.plot(pareto_random_op_10_df['waf'], label='pareto_random_op_10')
# ax.legend()

pareto_random_op_10_df.plot(y='waf', ax=ax, label='pareto_random_op_10='+str(pareto_random_op_10_df['waf'].max())
pareto_random_op_15_df.plot(y='waf', ax=ax, label='pareto_random_op_15='+str(pareto_random_op_15_df['waf'].max())
sequential_op_10_df.plot(y='waf', ax=ax, label='sequential_op_10='+str(sequential_op_10_df['waf'].max())
sequential_op_15_df.plot(y='waf', ax=ax, label='sequential_op_15='+str(sequential_op_15_df['waf'].max())
plt.show()
```
