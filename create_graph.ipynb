```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#  read json file into a pandas dataframe
pareto_random_op_5_df = pd.read_json('trace_files/graph_data/op_5_65_t/pareto_trace.log_waf.json')
pareto_random_op_10_df = pd.read_json('trace_files/graph_data/op_10_60_t/pareto_trace.log_waf.json')
pareto_random_op_15_df = pd.read_json('trace_files/graph_data/op_15_55_t/pareto_trace.log_waf.json')
pareto_random_op_20_df = pd.read_json('trace_files/graph_data/op_20_50_t/pareto_trace.log_waf.json')
sequential_op_5_df = pd.read_json('trace_files/graph_data/op_5_65_t/seq_trace.log_waf.json')
sequential_op_10_df = pd.read_json('trace_files/graph_data/op_10_60_t/seq_trace.log_waf.json')
sequential_op_15_df = pd.read_json('trace_files/graph_data/op_15_55_t/seq_trace.log_waf.json')
sequential_op_20_df = pd.read_json('trace_files/graph_data/op_20_50_t/seq_trace.log_waf.json')
```

```python
# plot the data and show the max value of waf for each trace

fig, ax = plt.subplots()
ax.set_title('WAF vs Time')
ax.set_xlabel('Sequence Number')
ax.set_ylabel('WAF')
pareto_random_op_5_df.plot(y='waf', ax=ax, label='pareto_random_op_5='+str(round(pareto_random_op_5_df['waf'].max(),2)))
pareto_random_op_10_df.plot(y='waf', ax=ax, label='pareto_random_op_10='+str(round(pareto_random_op_10_df['waf'].max(),2)))
pareto_random_op_15_df.plot(y='waf', ax=ax, label='pareto_random_op_15='+str(round(pareto_random_op_15_df['waf'].max(),2)))
pareto_random_op_20_df.plot(y='waf', ax=ax, label='pareto_random_op_20='+str(round(pareto_random_op_20_df['waf'].max(),2)))
sequential_op_5_df.plot(y='waf', ax=ax, label='sequential_op_5='+str(round(sequential_op_5_df['waf'].max(),2)))
sequential_op_10_df.plot(y='waf', ax=ax, label='sequential_op_10='+str(round(sequential_op_10_df['waf'].max(),2)))
sequential_op_15_df.plot(y='waf', ax=ax, label='sequential_op_15='+str(round(sequential_op_15_df['waf'].max(),2)))
sequential_op_20_df.plot(y='waf', ax=ax, label='sequential_op_20='+str(round(sequential_op_20_df['waf'].max(),2)))
plt.show()
```

```python
#  read json file into a pandas dataframe
pareto_random_op_5_cum_time_df = pd.read_json('trace_files/graph_data/op_5_65_t/pareto_trace.log_cummalative_time_per_packet.json')
pareto_random_op_10_cum_time_df = pd.read_json('trace_files/graph_data/op_10_60_t/pareto_trace.log_cummalative_time_per_packet.json')
pareto_random_op_15_cum_time_df = pd.read_json('trace_files/graph_data/op_15_55_t/pareto_trace.log_cummalative_time_per_packet.json')
pareto_random_op_20_cum_time_df = pd.read_json('trace_files/graph_data/op_20_50_t/pareto_trace.log_cummalative_time_per_packet.json')
sequential_op_5_cum_time_df = pd.read_json('trace_files/graph_data/op_5_65_t/seq_trace.log_cummalative_time_per_packet.json')
sequential_op_10_cum_time_df = pd.read_json('trace_files/graph_data/op_10_60_t/seq_trace.log_cummalative_time_per_packet.json')
sequential_op_15_cum_time_df = pd.read_json('trace_files/graph_data/op_15_55_t/seq_trace.log_cummalative_time_per_packet.json')
sequential_op_20_cum_time_df = pd.read_json('trace_files/graph_data/op_20_50_t/seq_trace.log_cummalative_time_per_packet.json')
```

```python
# plot the data and show the max value of waf for each trace

fig, ax = plt.subplots()
ax.set_title('Cumulative Time vs Write Packet')
ax.set_xlabel('Write Packet')
ax.set_ylabel('Cumulative Time')
pareto_random_op_5_cum_time_df.plot(y='time', ax=ax, label='pareto_random_op_5='+str(round(pareto_random_op_5_cum_time_df['waf'].max(),2)))
pareto_random_op_10_cum_time_df.plot(y='time', ax=ax, label='pareto_random_op_10='+str(round(pareto_random_op_10_cum_time_df['waf'].max(),2)))
pareto_random_op_15_cum_time_df.plot(y='time', ax=ax, label='pareto_random_op_15='+str(round(pareto_random_op_15_cum_time_df['waf'].max(),2)))
pareto_random_op_20_cum_time_df.plot(y='time', ax=ax, label='pareto_random_op_20='+str(round(pareto_random_op_20_cum_time_df['waf'].max(),2)))
sequential_op_5_cum_time_df.plot(y='time', ax=ax, label='sequential_op_5='+str(round(sequential_op_5_cum_time_df['waf'].max(),2)))
sequential_op_10_cum_time_df.plot(y='time', ax=ax, label='sequential_op_10='+str(round(sequential_op_10_cum_time_df['waf'].max(),2)))
sequential_op_15_cum_time_df.plot(y='time', ax=ax, label='sequential_op_15='+str(round(sequential_op_15_cum_time_df['waf'].max(),2)))
sequential_op_20_cum_time_df.plot(y='time', ax=ax, label='sequential_op_20='+str(round(sequential_op_20_cum_time_df['waf'].max(),2)))
plt.show()
```
